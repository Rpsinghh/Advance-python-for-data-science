# use of subplots  in plots

```python
plt.figure() # create a plot figure

# create the first of two panels and set current axis
plt.subplot(2,1,1) # rows,columns,panel number
plt.plot(x,np.sin(x)) 

# create second panel
plt.subplot(2,1,2)
plt.plot(x,np.cos(x))
```

# create a grid of plots

```python
# axis will be the axis of objects which you input
fig,ax=plt.subplots(3)

# call plot() for appropriate object
ax[0].plot(x,np.sin(x))
ax[1].plot(x,np.cos(x))
ax[2].plot(x+0.5,np.sin(x))
```

# SIMPLE LINE PLOTS

# ===================

```python
# this is use for inline function
%matplotlib inline
plt.style.use('seaborn-whitegrid') # setting another style
```

```python
fig=plt.figure() # size of figure
```

```python
ax=plt.axes() # axis x and y
```

```python
x=np.linspace(0,10,1000)
fig=plt.figure()
ax=plt.axes()

ax.plot(x,np.sin(x))
ax.plot(x,np.cos(x),linestyle=':')
```

```python
# Adjusting the line color and styles 
plt.plot(x,np.sin(x),color='blue') # specifying the color
plt.plot(x,np.cos(x),color='g') # short forms
plt.plot(x,np.sin(x-1),color='0.75') # grayscale 
plt.plot(x,np.cos(x-1),color='#FFDD44') # Hex code
plt.plot(x,np.sin(x-2),color=(1.0,0.2,0.4)) # RGB color
```

```python
plt.plot(x,np.sin(x),color='chartreuse') # html color
```

Linestyle

```python
plt.plot(x,x,linestyle='solid')
plt.plot(x,x+0.5,linestyle='dashed')
plt.plot(x,x+1.0,linestyle='dashdot')
plt.plot(x,x+1.5,linestyle='dotted')

# for short you can use
plt.plot(x,x+2,linestyle='-')
plt.plot(x,x+2.5,linestyle='--')
plt.plot(x,x+3,linestyle='-.')
plt.plot(x,x+3.5,linestyle=':')

# color and linestyle both
plt.plot(x, x + 4, '-g')  # solid green
plt.plot(x, x + 5, '--c') # dashed cyan
plt.plot(x, x + 6, '-.k') # dashdot black
plt.plot(x, x + 7, ':r');  # dotted red
plt.show()
```

```python
plt.plot(x,np.sin(x))

# set the x limit and y limit
plt.xlim(-1,11)
plt.ylim(-1.5,1.5)
```

```python
plt.plot(x,np.sin(x))

plt.xlim(10,0)
plt.ylim(1.2,-1.2)
```

```python
# xmin,xmax,ymin,ymax
plt.plot(x,np.cos(x))
plt.axis([-1,11,-1.5,1.5])
```

```python
# fix the axis by names
plt.plot(x,np.sin(x))
# plt.axis('tight')  # tight,on,off,equal,scaled,auto,image,square
# plt.axis('on')
plt.plot(x,np.cos(x))
# plt.axis('off')
# plt.axis('auto')
# plt.axis('image')
# plt.axis('square')
plt.axis('equal')


```

# Labeling plots

```python
plt.plot(x,np.sin(x))
plt.title(label='A Sin Function Curve',fontsize=18,fontweight=900,
          color='red',verticalalignment='baseline')
# fontsize': 'fontweight','color','verticalalignment': 
# ,'horizontalalignment'
```

```python
plt.plot(x,np.sin(x))
plt.xlabel('X',fontsize=20,fontweight=900,color='k',loc='center')
```

```python
# loc =center,top,bottom
plt.plot(x,np.sin(x))
plt.ylabel('SinX',fontsize=20,fontweight=900,color='Blue',loc='top')
plt.ylabel('SinX',fontsize=20,fontweight=900,color='Blue',loc='bottom')
```

```python
plt.plot(x,np.sin(x),'-g',label='sin(x)')
plt.plot(x,np.cos(x),':b',label='Cos(x)')
plt.axis('equal')
plt.legend() # legend will show you the information about everything
```

```python
# use all things in one
ax=plt.axes()

ax.plot(x,np.sin(x))
ax.plot(x,np.cos(x))

ax.set(xlim=(0,10),ylim=(-2,2),
      xlabel='x',ylabel='Sin(x)',title='A Simple plot')
```
